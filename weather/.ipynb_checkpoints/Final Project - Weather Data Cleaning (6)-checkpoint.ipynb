```python
# import packages
import numpy as np
import pandas as pd
```

```python
weather = pd.read_csv('complete_weather_data_1960-2018.csv')
weather.info()
```

```python
# make date the index and check the head
weather.set_index('DATE', inplace=True)
weather.index = pd.to_datetime(weather.index)
weather.head()
```

```python
# check the info
weather.info()
```

```python
# check # of empty values
weather.isnull().sum()
```

```python
# we delete the rows for the data that missing too much
dropThis = pd.date_range('1978-04-01','1984-01-01')
weather = weather[~weather.index.isin(dropThis)]
weather.dropna(subset = ['Snowfall','TMAX','TMIN'])
```

```python
# check the # of empty values
weather.isnull().sum()
```

```python
# fill the blanks by 0 and check the data
weather.fillna(value = 0, inplace = True) 
weather.isnull().sum()
```

```python
# for extreme weather
def extreme(df):
    if df['Heavy_fog'] == 1 or df['Thunder'] == 1 or df['Hail'] == 1 or df['High_winds'] == 1 or df['Freezing'] == 1 or df['Snow'] == 1:
        return 1
    else: 
        return 0
```

```python
weather['Extreme'] = weather.apply(extreme, axis=1)
```

```python
# output it as a csv and then check it 
weather.to_csv('complete_weather_data_1960-2018_cleaned.csv')
```
